2. Economía Neoinstitucionalista: Impacto de las Instituciones en el Crecimiento
(Enfoque en cómo las instituciones afectan el desarrollo económico)

```python
import pandas as pd
import seaborn as sns

# Datos simulados: Calidad institucional vs. PIB per cápita
data = {
    "País": ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"],
    "Calidad_Institucional": [3, 5, 7, 8, 4, 6, 9, 2, 1, 5],  # 1-10 (peor-mejor)
    "PIB_per_capita": [5000, 12000, 25000, 30000, 8000, 18000, 35000, 3000, 2000, 10000]  # USD
}

df = pd.DataFrame(data)

# Gráfico de correlación
plt.figure(figsize=(10, 6))
sns.regplot(x="Calidad_Institucional", y="PIB_per_capita", data=df, ci=None)
plt.title("Economía Neoinstitucionalista: Instituciones Fuertes → Mayor PIB")
plt.xlabel("Calidad Institucional (1-10)")
plt.ylabel("PIB per cápita (USD)")
plt.grid()
plt.show()
```

Explicación:
Este gráfico muestra cómo países con mejores instituciones (leyes claras, derechos de propiedad, baja corrupción) tienden a tener mayor PIB per cápita. Los gobiernos influenciados por esta teoría promueven reformas institucionales para el desarrollo.

3. Economía Conductual: Sesgos en Decisiones de Política Pública
(Enfoque en cómo la psicología afecta decisiones económicas)

```python
# Ejemplo de "nudge" (empujón conductual)
choices = ["Opción A (Energía Limpia)", "Opción B (Energía Contaminante)"]
default_effect = [0.75, 0.25]  # 75% eligen A cuando es opción por defecto

# Gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(choices, default_effect, color=["green", "red"])
plt.title("Economía Conductual: Efecto del 'Nudge' en Decisiones Verdes")
plt.ylabel("Proporción de Elección")
plt.ylim(0, 1)
plt.show()
```

Explicación:
Este gráfico ilustra cómo pequeños cambios en el diseño de políticas ("nudges") pueden influir en decisiones ciudadanas (ej: energía limpia por defecto). Gobiernos conductuales usan esto en pensiones, salud y sostenibilidad.

Modelo de Solow Básico
(Enfoque neoclásico: capital, trabajo y tecnología determinan el crecimiento)

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del modelo
A = 1.0       # Tecnología (TFP)
K = np.linspace(0.1, 100, 100)  # Capital per cápita
alpha = 0.3   # Participación del capital en la producción
delta = 0.1   # Tasa de depreciación
s = 0.2       # Tasa de ahorro
n = 0.02      # Tasa de crecimiento poblacional

# Función de producción Cobb-Douglas
def production(K, A, alpha):
    return A * (K ** alpha)

# Inversión y depreciación
Y = production(K, A, alpha)
investment = s * Y
depreciation = (delta + n) * K

# Punto de equilibrio (steady-state)
ss_idx = np.argwhere(np.diff(np.sign(investment - depreciation)))[0][0]
ss_k = K[ss_idx]
ss_y = Y[ss_idx]

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(K, Y, label="Producción ($Y = A \cdot K^\\alpha$)", color="blue")
plt.plot(K, investment, label="Inversión ($sY$)", linestyle="--", color="green")
plt.plot(K, depreciation, label="Depreciación ($(\\delta + n)K$)", linestyle=":", color="red")
plt.axvline(x=ss_k, linestyle="-.", color="gray", label=f"Equilibrio ($K^* = {ss_k:.1f}$)")
plt.scatter(ss_k, ss_y, color="black", zorder=5)
plt.title("Modelo de Solow: Crecimiento Económico")
plt.xlabel("Capital per cápita ($K$)")
plt.ylabel("Producción per cápita ($Y$)")
plt.legend()
plt.grid()
plt.show()
```
